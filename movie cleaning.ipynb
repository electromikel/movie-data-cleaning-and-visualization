```python
titanic.columns = titanic.columns.str.lower()
```

```python
import numpy as np 
```

```python
titanic['faith'] = np.where(titanic['survived']== 0, 'died', 'survived')
```

```python
titanic
```

```python
titanic['pcclass'] = titanic['pclass']. replace(3, 'third class')
titanic['pcclass'] = titanic['pclass']. replace(2, 'second class')
titanic['pcclass'] = titanic['pclass']. replace(1, 'first class')
```

```python
titanic
```

```python
titanic.age = titanic.age.fillna(0)
```

```python
titanic.age = titanic.age.astype(int)
```
